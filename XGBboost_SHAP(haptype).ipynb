```python
import pandas as pd
import numpy as np
import xgboost
import shap
from sklearn.model_selection import train_test_split
import matplotlib.pylab as pl
import xgboost as xgb
from sklearn.metrics import mean_squared_error
# print the JS visualization code to the notebook
shap.initjs()
```

```python
### Zm00001d008941
```

```python
Zm00001d008941 = pd.read_csv("./Zm00001d008941/shape.csv")
Zm00001d008941_data = Zm00001d008941.iloc[:, 2:]
Zm00001d008941_label = Zm00001d008941.iloc[:, 1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(Zm00001d008941_data, Zm00001d008941_label, test_size=0.2, random_state=42)
```

```python
model = xgb.XGBRegressor(objective ='reg:squarederror', seed=42,
                         colsample_bytree=1,
                         gamma=0.6,
                         learning_rate=0.1,
                         max_depth=5,
                         min_child_weight=5,
                         n_estimators=200,
                         subsample=0.8).fit(X_train, y_train)

explainer = shap.Explainer(model)

shap_values = explainer(X_test)

shap.plots.bar(shap_values, max_display=30, show=False)



```
